## Bar plot

```python
nut_butter_price={
    "Almond Butter": 10,
    "Peanut Butter": 8,
    "Cashew Butter": 12
}

fig, ax = plt.subplots()
ax.set(title='Nut Buttter Price', ylabel='Price', xlabel='Type')
ax.bar(nut_butter_price.keys(), nut_butter_price.values());
```

## Histogram

```python
x= np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

# Subplots

```python
# option 1
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax1.plot(x,x);
ax2.scatter(np.random.random(10),np.random.randn(10));
ax3.bar(nut_butter_price.keys(), nut_butter_price.values());
ax4.hist(np.random.random(1000));
```

```python
# option 2
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(x,x);
ax[0,1].scatter(np.random.random(10),np.random.randn(10));
ax[1,0].bar(nut_butter_price.keys(), nut_butter_price.values());
ax[1,1].hist(np.random.random(1000));
```

### Ploting From Pandas Dataframe

```python
car= pd.read_csv('car-sales.csv')
```

```python
car["Price"] = car['Price'].str.replace('[\$\,\.]', '', regex=True)
car
```

```python
car["Price"] = car['Price'].str[:-2]
```

```python
car['Price'].astype(int)
```

```python
car['Sale Date'] = pd.date_range('1/1/2022' , periods= 10)
car
```

```python
car['Total'] = car['Price'].astype(int).cumsum()
car
```

```python
car.plot(x='Sale Date', y='Total');
```

```python
car['Price'] = car['Price'].astype(int)
car.plot(x='Odometer (KM)', y='Price', kind='scatter');
```

```python
car['Odometer (KM)'].plot.hist(bins = 10);
```
